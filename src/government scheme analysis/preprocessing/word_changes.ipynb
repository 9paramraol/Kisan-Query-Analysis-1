```python
# The maping part

eline =  open('distinct_govt_query', 'r').read().splitlines()
hline =  open('../hindi_tranlations/nhindi.txt', 'r').read().splitlines()
all_lines = open('all_govt_query', 'r').read().splitlines()

e_h_trans = {}
for i in range(0,len(eline)):
    e_h_trans[eline[i]] = hline[i]
```

```python
#finding words before yojana
cnt = 0
dist_yojana = {}
for line in all_lines:
    try:
        wrds = shlex.split(e_h_trans[line])
        ind = wrds.index("योजना")
        sind = max(0,ind-2)
        y_str = ' '.join([str(elem) for elem in wrds[sind:ind]])
        if (y_str not in dist_yojana):
            dist_yojana[y_str] = 0 
        dist_yojana[y_str] += 1
        cnt += 1
    except:
        continue
```

```python
freq_yojana = [[100*dist_yojana[y]/cnt,y] for y in dist_yojana]
freq_yojana.sort()
freq_yojana.reverse()
```
