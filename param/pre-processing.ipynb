```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re
import shlex

def init():
    extra_stop_words_list = ['^ask.*', '^control', '^treatment', '^tell']
    extra_stop_words.update(extra_stop_words_list)
  
def filter_sentence(sentence):
    sentence = re.sub(r'[^a-zA-Z0-9 ]+', ' ', sentence)
    word_tokens = word_tokenize(sentence)  
    
    # sentence = [ps.stem(w) for w in word_tokens]
    sentence = word_tokens
    sentence = [w for w in sentence if not w in stop_words]
    sentence = [w for w in sentence if all(not re.search(x, w) != None for x in extra_stop_words)]
    sentence = ' '.join(sentence)

    return sentence    

stop_words = set(stopwords.words('english'))
newStopWords = ['technical','details','information','registration','farmer','query','relative','relative']
newStopWords += ['ongoing','survey','check','regarding','farmers','problem','problems','number','related','wants']
newStopWords += ['contact','know','want','hai','government','ki','jankari','ke','district','status','application']
newStopWords += ['meri','crop','agriculture','department','insurance','ka','block','wheat','paddy','agricultural','give','seed']
newStopWords += ['needs', 'weather', 'free', 'toll', 'get',  'officer',  'se', 'farm', 'need', 'de', 'date', 'u2026', 'nahi']
newStopWords += ['installment', 'general',  'animal', 'yellowing','ki']
for wrd in newStopWords:
    if( wrd not in stop_words):
        stop_words.add(wrd)

extra_stop_words = set()
ps = PorterStemmer()
init()
```

```python
file1 = open('idistinct.txt', 'r') 
Lines = file1.readlines()
```

```python
ans_list = []
for line in Lines:
    nline = filter_sentence(line)
    ans_list.append(nline)
```

```python
final_list = []
line_set = {""}
for line in ans_list:
    if not(line in line_set):
        final_list.append(line)
        line_set.add(line)
```

```python
corpus_set = {}

for line in final_list:
    try:
        words = shlex.split(line)
    except:
        continue
    for wrd in words:
        if(wrd not in corpus_set):
            corpus_set[wrd] = 0

freq_list = []
lc = 0

for line in final_list:
    for wrd in shlex.split(line):
        corpus_set[wrd] += 1
        
for wrd in corpus_set:
    freq_list.append((corpus_set[wrd],wrd))

freq_list.sort()
freq_list.reverse()

len(freq_list)
```

```python
with open('distinct.txt', 'w') as f:
    for item in final_list:
        f.write("%s\n" % item)
```
